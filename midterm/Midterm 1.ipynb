```python
import numpy as np
import pandas as pd
import scipy as sp
from sklearn.cluster import KMeans
import scipy.spatial.distance as dist
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.neighbors import kneighbors_graph
from scipy.spatial import distance_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#from plotly import express as px
```

```python
data = np.array([[2,2],
                [-1,1],
                [3,1],
                [0,-1],
                [-2,-2]])

centroids = np.array([[-3,-1],
                     [2,1]])

dist_matrix = np.zeros((5,2))

def k_mean_euclidean(data,centroids,dist_matrix):
    for i in range(data.shape[0]):
        for j in range(centroids.shape[0]):
            distance = dist.euclidean(data[i],centroids[j])# euclidean distance
            dist_matrix[i,j] = distance
    cluster = np.asarray([np.argmin(i) for i in dist_matrix])
    new_data = pd.DataFrame(data)
    new_data["cluster"] = cluster
    centroids = new_data.groupby(['cluster']).mean().to_numpy() # calculate the new center for each centroid

    
    return centroids,cluster,dist_matrix

centroids,cluster,dist_matrix = k_mean_euclidean(data,centroids,dist_matrix)

print ("Iteration 1:")
print ("#################")
print ("centroids:\n", centroids) # centroids coordinates
print ("#################")
print ("cluster:\n", cluster) # cluster assignment
print ("#################")
print ("distance matrix:")
print (dist_matrix)
print ("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@\n")


centroids,cluster,dist_matrix = k_mean_euclidean(data,centroids,dist_matrix)

print ("Iteration 2:")
print ("#################")
print ("centroids:\n", centroids) # centroids coordinates
print ("#################")
print ("cluster:\n", cluster) # cluster assignment
print ("#################")
print ("distance matrix:")
print (dist_matrix)
print ("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@\n")

centroids,cluster,dist_matrix = k_mean_euclidean(data,centroids,dist_matrix)

print ("Iteration 3:")
print ("#################")
print ("centroids:\n", centroids) # centroids coordinates
print ("#################")
print ("cluster:\n", cluster) # cluster assignment
print ("#################")
print ("distance matrix:")
print (dist_matrix)
print ("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@\n")

centroids,cluster,dist_matrix = k_mean_euclidean(data,centroids,dist_matrix)

print ("Iteration 4:")
print ("#################")
print ("centroids:\n", centroids) # centroids coordinates
print ("#################")
print ("cluster:\n", cluster) # cluster assignment
print ("#################")
print ("distance matrix:")
print (dist_matrix)
print ("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@\n")
```

```python
from sklearn.cluster import KMeans
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.neighbors import kneighbors_graph
import matplotlib.pyplot as plt 

data = np.array([[-1,2],
                [-1,1],
                [-1,0],
                [-1,-1],
                [-1,-2],
                [0,0],
                 [1,0],
                 [2,0]])

data = pd.DataFrame(data)
data = data.rename(columns = {0:"x",1:"y"})

#calculate the distance to its two nearest neighbors, obtain a distance matrix
z = np.around(kneighbors_graph(data,n_neighbors=2,mode='distance').toarray(),2)

# make the distance matrix undirected.
for i in range(z.shape[0]):
    for j in range(z.shape[0]):
        if z[i,j] !=0 and z[j,i]==0:
            z[j,i]=z[i,j]


# get the adjancency matrix based on euclidean distance and two nearest neighbors
A = np.exp(-z)
A= np.where(A==1,0, A)

# get degree matrix
D = np.diag(A.sum(axis=0))

# get laplacian matrix
L = D-A
L_pd = pd.DataFrame(L)
print ("Laplacian matrix:")
display(L_pd)

# find the eigenvalues and eigenvectors
vals, vecs = np.linalg.eig(L)

# sort
vecs = vecs[:,np.argsort(vals)]
vals = vals[np.argsort(vals)]

kmeans = KMeans(n_clusters=2)
kmeans.fit(vecs[:,0:2])
colors = kmeans.labels_
data['color'] = colors
sns.lmplot(x='x',y='y',hue="color", data=data,fit_reg=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import rescale, resize, downscale_local_mean
import os
print (os.getcwd())
os.chdir("yalefaces/")
# define a functio to vectorize the image data
def vectorize(image_name):
    image = plt.imread(image_name)
    image = downscale_local_mean(image, (4,4))
    image_vectorized = image.flatten("C")
    return image_vectorized

def eigenface(subject_name):
    # get the eigenface for subject
    sub14 = [os.listdir()[i] for i in range(len(os.listdir())) if os.listdir()[i].startswith(subject_name)]

    for i in range(len(sub14)):
        if i ==0:
            n = vectorize(sub14[i])# get the first vectorized image data
        else:
            n = np.column_stack((n,vectorize(sub14[i]))) # add the rest of the image data

    new = n.T.dot(n) # compute the X^TX matrix

    # get the eigenvalues and eigenvectors
    vals, vecs =  np.linalg.eig(new)

    # sort them from high to low
    vecs = -vecs[:,np.argsort(-vals)]
    vals = vals[np.argsort(-vals)]
   

    print ("The top 6 eigenfaces of",subject_name)
    for i in range(6):
        face = n.dot(vecs[:,i]) # compute the eigenface
        face_reshape = np.reshape(face, (61,80)) # reshape the data to its original form
        plt.subplot(2,3,i+1)
        plt.title('Eigenface_'+str(i))
        plt.imshow(face_reshape,cmap="Greys_r")
        plt.axis('off')
    face_1 = n.dot(vecs[:,0])
    return face_1
face_01 = eigenface("subject01")
```

```python
import numpy as np
import networkx as nx
from sklearn.neighbors import kneighbors_graph
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


def Matrix_D(W):
    # Generate Graph and Obtain Matrix D, \\
    # from weight matrix W defining the weight on the edge between each pair of nodes.
    # Note that you can assign sufficiently large weights to non-existing edges.

    n = np.shape(W)[0]
    Graph = nx.DiGraph()
    for i in range(n):
        for j in range(n):
            Graph.add_weighted_edges_from([(i,j,min(W[i,j], W[j,i]))])
    
    res = dict(nx.all_pairs_dijkstra_path_length(Graph))
    D = np.zeros([n,n])
    for i in range(n):
        for j in range(n):
            D[i,j] = res[i][j]
    np.savetxt('D.csv', D)
    return D
```

```python
data = pd.read_csv("isomap.csv",header=None) # read in the file

# get the pair wise distance matrix of 100 closest points
W = kneighbors_graph(data.T,n_neighbors=100,mode="distance").toarray() 

A = kneighbors_graph(data.T,n_neighbors=100,include_self=False).toarray() # adjacency matrix

plt.figure(figsize=(15,10))
sns.heatmap(A)# visulize the adjancency matrix as a heatmap
plt.show()

W[W==0]=8200 # assign the non-existing edge a relative large number
D = Matrix_D(W) # calculating pairwise shortest distance matrix

m = data.shape[1] # number of images
I = np.identity(m) # the m by m identity matrix 
one_vector = np.ones((m,m)) # m by m one matrix
H = I - (1/m)*one_vector # get the centering matrix
C = (-1/(2*m))*(H.dot(np.square(D)).dot(H)) #distance matrix C
```

```python
vals, vecs = np.linalg.eig(C)

# sort them from high to low
vals = vals[np.argsort(-vals)]
vecs = vecs[:,np.argsort(-vals)]

# get two leading eigen vectors and eigenvalues
leading_vecs = vecs[:,0:2].real
leading_vals = vals[0:2].real

# construct the scaling matrix with eigenvalues
scaling_matrix = np.zeros((2,2))

for i in range(2):
    for j in range(2):
        if i == j:
            scaling_matrix[i,j] = np.sqrt(leading_vals[i])
            
# get the first two components of ISOMAP
Z_t = leading_vecs.dot(scaling_matrix)
Z_t = pd.DataFrame(Z_t)
Z_t = Z_t.rename(columns= {0:'x',1:"y"})
Z_t["image_no"] = np.array([i for i in range(data.shape[1])])


plt.figure(figsize=(15,10))
p1 = sns.regplot(data=Z_t, x="x",y="y",fit_reg=False)
image_list= [691,140,154, 406,631,278]

# annotate data points
#p1.text(Z_t["x"][image_list[0]],Z_t["y"][image_list[0]],Z_t["image_no"][image_list[0]],size='medium', color='black', weight='semibold')
#p1.text(Z_t["x"][image_list[1]],Z_t["y"][image_list[1]],Z_t["image_no"][image_list[1]],size='medium', color='black', weight='semibold')
#p1.text(Z_t["x"][image_list[2]],Z_t["y"][image_list[2]],Z_t["image_no"][image_list[2]],size='medium', color='black', weight='semibold')


for i in image_list:
    p1.text(Z_t["x"][i],Z_t["y"][i],Z_t["image_no"][i],size='medium', color='black', weight='semibold')

plt.show()
plt.figure(figsize=(15,10))
for i in image_list:
        face_reshape = np.reshape(np.asarray(data.iloc[:,i]), (64,64)).T # reshape the data to its original form
        plt.subplot(2,len(image_list)/2,image_list.index(i)+1)
        plt.title('Eigenface_'+str(i))
        plt.imshow(face_reshape)
        plt.axis('off')
```
